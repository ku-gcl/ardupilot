```python
import pandas as pd
```

```python
f = open('0_LogStructure_20221104.txt', 'r')
data = f.read()
f.close()
```

LoggerMessage, Description, Field, URL

```python
data_list = data.split('\n')
```

```python
# copterのlogに含まれるパラメータ
copter_param = "AHR2 ARM ATT BARO BAT CMD CTRL CTUN D32 DSF DU32 ERR EV FILE FMT FMTU IMU IOMC MAG MODE MOTB MSG MULT PARM PM POWR PSCD PSCE PSCN RATE RCI2 RCIN RCOU STAK UNIT VER VIBE XKF1 XKF2 XKF3 XKF4 XKF5 XKFS XKQ XKT XKV1 XKV2"
copter_log = copter_param.split(" ")
print(copter_log)
```

```python
result_list = []
Filename = 0; LoggerMessage = 0; Description = 0; 
Field = 0; FieldName = 0; FieldDescription = 0;
Url = 0;

for i, value in enumerate(data_list):
    if "##" in value:
        Filename = value.split("##")[1]
        continue
    if "LoggerMessage" in value:
        LoggerMessage = value.split(": ")[1]
        if LoggerMessage in copter_log:
            isCopter = 1
        else:
            isCopter = 0
        continue
    if "Description" in value:
        Url = ""        # initialize URL
        Description = value.split(": ")[1]
        continue
    if "Field" in value:
        Field = value.split(": ")
        FieldName = Field[1]
        FieldDescription = Field[2]
    if "URL" in value:
        Url = value.split(": ")[1]
    result_list.append([Filename, LoggerMessage, Description, Url, FieldName, FieldDescription, isCopter])
        
    
```

```python
# columns = ["Filename", "LoggerMessage", "Description", "FieldName", "FieldDescription", "Url"]
columns = ["Filename", "LoggerMessage", "Description", "Url", "FieldName", "FieldDescription", "isCopter"]
df = pd.DataFrame(result_list, columns=columns)
df.head(20)
```

```python
df.to_excel("LOG_PARAMETER.xlsx")
```
